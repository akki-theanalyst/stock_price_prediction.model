```python
import yfinance as yf
import pandas as pd
```

```python
# Download historical data
tickers = ['HDFCBANK.NS', 'SBIN.NS', 'ICICIBANK.NS']
stock_data = yf.download(tickers, start='2022-05-30', end = '2025-05-30', group_by ='ticker')


#Check data for Sbi bank

hdfc_bank_data = stock_data['HDFCBANK.NS']
sbi_data = stock_data['SBIN.NS']
icici_bank_data = stock_data['ICICIBANK.NS']

print(sbi_data.head())
```

```python
# Specify the file path where the CSV should be saved.

file_path1 = 'hdfc_bank_data.csv'
file_path2 = 'sbi_data.csv'
file_path3 = 'icici_bank_data.csv'

# Save the dataframe to a csv file.
hdfc_bank_data.to_csv(file_path1, index=True)
sbi_data.to_csv(file_path2, index=True)
icici_bank_data.to_csv(file_path3, index=True)
```
